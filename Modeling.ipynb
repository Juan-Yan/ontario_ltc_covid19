```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import data, split independent variables and dependent variables
dataset = pd.read_csv('dataset.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
dataset.head()
```

```python
print(x)
print(y)
```

```python
# split train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0 )
```

```python
# training the multiple linear regression model on the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Predicting the test set results
y_pred = regressor.predict(x_test)
np.set_printoptions(precision = 2) # set decimal to 2
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),axis = 1)) # reshape(# rows, # cols)
# compare the predicted result with y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# Random Forest
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state=0)
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
dataset = pd.read_csv('dataset.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
y = y.reshape(len(y),1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_x.fit_transform(x_train)

sc_y = StandardScaler()
sc_y.fit_transform(y_train)
```

```python
x_train = sc_x.transform(x_train)
y_train = sc_y.transform(y_train)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(x_train,y_train)
```

```python
y_pred = sc_y.inverse_transform(regressor.predict(sc_x.transform(x_test)))
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
